```julia
function amp(z,x0)
        k = 5
        r(x) = sqrt( z^2  + (x - x0)^2 )
        int, err = quadgk(x -> exp( 1im * k * r(x) ), -1, 1)
        int
end
```

```julia
for z in 0:0.1:3
   for x0 in -2:0.1:2
        amp(z,x0)
    end
end
```

```julia
x0s = -2:0.1:2
zs = 0:0.1:3
intensity = [abs(amp(z,x0))^2 for x0=x0s, z=zs]
```

```julia
using Plots
```

```julia
l = @layout [a{0.7w} b; c{0.2h}]
p = plot(zs, x0s, intensity, st = :surface, layout=l)
plot!(p[1], camera = (10 * (1 + cos(0)), 40))
```
