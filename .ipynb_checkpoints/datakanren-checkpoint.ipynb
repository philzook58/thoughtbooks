```python
from dataclasses import dataclass
from typing import Any

@dataclass
class Plus:
    a : Any
    b : Any
        
@dataclass
class Lit:
    n : int
        
#@dataclass
#class PatVar:
#    name: str

class PatternFailure(Exception):
    """Base class for exceptions in this module."""
    pass
        
def lit(n):
    def res(t,b):
        if isinstance(t, Lit):
            if t.n == n:
                return b
            else:
                raise PatternFailure
    return res

def var(name):
    def res(t,bind):
        if name in bind:
            if bind[name] == t:
                return bind
            else:
                raise PatternFailure
        else:
            bind[name] = t
            return bind
    return res

def plus(pa,pb):
    def res(t,bind):
        if isinstance(t, Plus):
            bind = pa(t.a, bind)
            bind = pb(t.b, bind)
            return bind
        else:
             raise PatternFailure
    return res

def match(pat,t):
    try:
        bind = {}
        return pat(t,bind)
    except PatternFailure:
        return None

print(match(lit(2), Lit(3)))
print(match(lit(3), Lit(3)))
print(match(var("x"), Lit(3)))
print(match(plus(lit(2), var("x")), Plus(Lit(2),Lit(6))))
print(match(plus(var("x"), var("x")), Plus(Lit(2),Lit(6))))
print(match(plus(var("x"), var("x")), Plus(Lit(6),Lit(6))))
```
