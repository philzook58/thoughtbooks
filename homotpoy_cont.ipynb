```julia
using HomotopyContinuation
fresh = 0

function var(n)
    global fresh += 1
    variables("var$fresh", 1:n)
end

# (inputs variables, outputs variables, equations)
id(n) = () -> begin
    x = var(n) 
   #@unique_var x[1:n] # this is giving me trouble for some reason? macro expansion at the wrong time?
   (x, x, x-x)
end
fst(a,b) = () -> begin
    x = var(a)
    y = var(b)
    ([x ; y], x, [x-x ; y-y])
end
snd(a,b) = () -> begin
    x = var(a)
    y = var(b)
    ([x ; y], y, [x-x ; y-y])
end


compose(f,g) = () -> begin
     (x,y,cf) = f()
     (y1,z,cg) = g()
     (x, z, [cf ; cg ; y - y1])
end

converse(f) = () -> begin
     (x,y,eq) = f()
     (y,x,eq)
end

fst(1,1)()
id(3)()
compose(id(3), id(3))()

run(f) = begin
    (ins, outs, eqs) = f()
    params = [ins ; outs] # maybe params = ins ? We want 0d queries.
    #vars = variables(eqs, parameters = params)
    #println(eqs)
    System(eqs; parameters = params)
end
run(compose(id(3), id(3)))
```
